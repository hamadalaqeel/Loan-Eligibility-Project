```python
debt_Consolidation.dropna(inplace = True)
```

```python
debt_Consolidation['int_rate'] = debt_Consolidation.int_rate.astype(int)
debt_Consolidation['installment'] = debt_Consolidation.int_rate.astype(int)
debt_Consolidation['emp_length'] = debt_Consolidation.int_rate.astype(int)
debt_Consolidation['annual_inc'] = debt_Consolidation.int_rate.astype(int)
debt_Consolidation['int_rate'] = debt_Consolidation.int_rate.astype(int)
debt_Consolidation['int_rate'] = debt_Consolidation.int_rate.astype(int)
```

```python
debt_Consolidation.info()
```

```python
debt_Consolidation.drop('purpose', axis=1, inplace=True)
```

```python
debt_Consolidation.head()
```

```python
debt_Consolidation.drop('Column', axis=1, inplace=True)
```

```python
debt_Consolidation.head()
```

```python
labels_loanStatus = debt_Consolidation['loan_status'].astype('category').cat.categories.tolist()
replace_map_loan_status = {'loan_status' : {k: v for k,v in zip(labels_loanStatus,list(range(1,len(labels_loanStatus)+1)))}}
print(replace_map_loan_status)
```

```python
debt_Consolidation.replace(replace_map_loan_status, inplace=True)
```

```python
labels_addr_state = debt_Consolidation['addr_state'].astype('category').cat.categories.tolist()
replace_map_comp_addr_state = {'addr_state' : {k: v for k,v in zip(labels_addr_state,list(range(1,len(labels_addr_state)+1)))}}
print(replace_map_comp_addr_state)
```

```python
debt_Consolidation.replace(replace_map_comp_addr_state, inplace = True)
```

```python
debt_Consolidation.head()
```

```python

y=debt_Consolidation['int_rate']
X = debt_Consolidation.drop('int_rate', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25,random_state=101)
```

```python
X_train.shape
```

```python
X_test.head()
```

```python
y_train.head()
```

```python
y_test.shape
```

```python
# reference: https://towardsdatascience.com/random-forest-in-python-24d0893d51c0
def pred_score(model):
    pred = model.predict(X_test)
    # Calculate the absolute errors
    errors = abs(pred - y_test)
    # Calculate mean absolute percentage error
    mape = 100 * (errors / y_test)
    accuracy = 100 - np.mean(mape)
    return round(accuracy, 2)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
          intercept_scaling=1, max_iter=110, multi_class='ovr', n_jobs=1,
          penalty='l1', random_state=None, solver='liblinear', tol=0.0001,
          verbose=0, warm_start=False)

logreg.fit(X_train, y_train)
print(f'Accuracy of Logistic regression classifier on training set: {round(logreg.score(X_train, y_train)*100,2)}%.')
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

knn.fit(X_train, y_train)
print(f'Accuracy of K-NN classifier on training set: {round(knn.score(X_train, y_train)*100,2)}%.')
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()

dt.fit(X_train, y_train)
print(f'Accuracy of Decision Tree classifier on training set: {round(dt.score(X_train, y_train)*100,2)}%.')
```

```python
from sklearn.svm import SVC

svm = SVC()

svm.fit(X_train, y_train)
print(f'Accuracy of SVM classifier on training set: {round(svm.score(X_train, y_train)*100,2)}%.')
```
